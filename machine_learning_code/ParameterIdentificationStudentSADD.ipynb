```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np
from scipy.stats import ttest_ind
```

```python
labelList = ["phq9", "q9"]
featureElist = ["Chi","PCA"]
modelTypelist =  [["NB"], ["XG2", "XG3", "XG4"], ["LR1", "LR2"], ["SVC1", "SVC2", "SVC3", "SVC4"], ["kNN1", "kNN3", "kNN5", "kNN7", "kNN9"], ["RF2g", "RF2e", "RF3g", "RF3e", "RF4g", "RF4e"]]
namelist = ["ScriptedText", "UnscriptedText", "Scripted", "Unscripted", "Text", "Transcript"]

scorelist = ["F1","Accuracy"]

namelist2 = []
labelList2 = []
featureElist2 = []
parameterlist2 = []
modelTypelist2 = [] #not going to be an arrary, choose 1 value per array
averageslist2 = []

for name in namelist:     
    for label in labelList:
        
        plist2 = []
        alist2 = []
        elist2 = []
        mlist2 = []

        for modelType in modelTypelist:
            
            plist = []
            alist = []
            elist = []
            
            for model in modelType:
                                
                #for featureE in featureElist:
                #for score in scorelist:
                filename = "resultsAll85/parameters" + name + label + ".csv"

                resultsDF = pd.read_csv(filename).fillna(0)
                resultsDFm = resultsDF[resultsDF["model"] == model]
                
                resultsDF1 = resultsDFm[resultsDFm["Engineering"] == featureElist[0]]
                resultsDF2 = resultsDFm[resultsDFm["Engineering"] == featureElist[1]]

                p1 = 0
                p2 = 0
                
                a1 = 0
                a2 = 0
                
                for parameter in sorted(set(resultsDF.nFeatures)):
                
                    list1a = resultsDF1[resultsDF1["nFeatures"] == parameter][scorelist[1]]
                    list1f = resultsDF1[resultsDF1["nFeatures"] == parameter][scorelist[0]]
                    #mean1 = (sum(list1a) + sum(list1f))/20
                    mean1 = sum(list1f)/10
                    
                    
                    list2a = resultsDF2[resultsDF2["nFeatures"] == parameter][scorelist[1]]
                    list2f = resultsDF2[resultsDF2["nFeatures"] == parameter][scorelist[0]]
                    #mean2 = (sum(list2a) + sum(list2f))/20
                    mean2 = sum(list2f)/10
                    
                    if parameter == sorted(set(resultsDF.nFeatures))[0]:
                        p1 = parameter
                        p2 = parameter
                        a1 = mean1
                        a2 = mean2
                    
                    
                    else:
                        if mean1 > a1:
                            #if (ttest_ind(list1a, resultsDF1[resultsDF1["nFeatures"] == p1][scorelist[1]])[1] <= 0.05) | (ttest_ind(list1f, resultsDF1[resultsDF1["nFeatures"] == p1][scorelist[0]])[1] <= 0.05):
                            if (ttest_ind(list1f, resultsDF1[resultsDF1["nFeatures"] == p1][scorelist[0]])[1] <= 0.05):

                                a1 = mean1
                                p1 = parameter
                                
                        if mean2 > a2:
                            #if (ttest_ind(list2a, resultsDF2[resultsDF2["nFeatures"] == p2][scorelist[1]])[1] <= 0.05) | (ttest_ind(list2f, resultsDF2[resultsDF2["nFeatures"] == p2][scorelist[0]])[1] <= 0.05):
                            if (ttest_ind(list2f, resultsDF2[resultsDF2["nFeatures"] == p2][scorelist[0]])[1] <= 0.05):

                                a2 = mean2
                                p2 = parameter
                    
                if p1 == p2:
                    plist.append(p1)
                    if a1 > a2:
                        alist.append(a1)
                        elist.append(featureElist[0])
                    elif a2 > a1:
                        alist.append(a2)
                        elist.append(featureElist[1])
                    else:
                        alist.append(a1)
                        elist.append("both")
                elif p1 < p2:
                    if (a2 > a1):
                        #if (ttest_ind(resultsDF1[resultsDF1["nFeatures"] == p1][scorelist[1]], resultsDF2[resultsDF2["nFeatures"] == p2][scorelist[1]])[1] <= 0.05) | (ttest_ind(resultsDF1[resultsDF1["nFeatures"] == p1][scorelist[0]], resultsDF2[resultsDF2["nFeatures"] == p2][scorelist[0]])[1] <= 0.05):
                        if (ttest_ind(resultsDF1[resultsDF1["nFeatures"] == p1][scorelist[0]], resultsDF2[resultsDF2["nFeatures"] == p2][scorelist[0]])[1] <= 0.05):

                            plist.append(p2)
                            alist.append(a2)
                            elist.append(featureElist[1]) 
                        else:
                            plist.append(p1)
                            alist.append(a1)
                            elist.append(featureElist[0])
                    else:
                        plist.append(p1)
                        alist.append(a1)
                        elist.append(featureElist[0])
                elif p2 < p1:
                    if (a1 > a2):
                        #if (ttest_ind(resultsDF1[resultsDF1["nFeatures"] == p1][scorelist[1]], resultsDF2[resultsDF2["nFeatures"] == p2][scorelist[1]])[1] <= 0.05) | (ttest_ind(resultsDF1[resultsDF1["nFeatures"] == p1][scorelist[0]], resultsDF2[resultsDF2["nFeatures"] == p2][scorelist[0]])[1] <= 0.05):
                        if (ttest_ind(resultsDF1[resultsDF1["nFeatures"] == p1][scorelist[0]], resultsDF2[resultsDF2["nFeatures"] == p2][scorelist[0]])[1] <= 0.05):

                            plist.append(p1)
                            alist.append(a1)
                            elist.append(featureElist[0]) 
                        else:
                            plist.append(p2)
                            alist.append(a2)
                            elist.append(featureElist[1])
                    else:
                        plist.append(p2)
                        alist.append(a2)
                        elist.append(featureElist[1])
                        
            
            p = plist[0]
            a = alist[0]
            e = elist[0]
            m = modelType[0]
            
            for i in range(1, len(modelType)):
                if plist[i] == p:
                    if alist[i] > a:
                        p = plist[i]
                        a = alist[i]
                        e = elist[i]
                        m = modelType[i] 
                elif plist[i] > p:
                    if alist[i] > a:
                        #if (ttest_ind(resultsDF[(resultsDF["nFeatures"] == p) & (resultsDF["Engineering"] == e) & (resultsDF["model"] == m)][scorelist[1]], resultsDF[(resultsDF["nFeatures"] == plist[i]) & (resultsDF["Engineering"] == elist[i]) & (resultsDF["model"] == modelType[i])][scorelist[1]])[1] <=0.05) | (ttest_ind(resultsDF[(resultsDF["nFeatures"] == p) & (resultsDF["Engineering"] == e) & (resultsDF["model"] == m)][scorelist[0]], resultsDF[(resultsDF["nFeatures"] == plist[i]) & (resultsDF["Engineering"] == elist[i]) & (resultsDF["model"] == modelType[i])][scorelist[0]])[1] <=0.05):
                        if (ttest_ind(resultsDF[(resultsDF["nFeatures"] == p) & (resultsDF["Engineering"] == e) & (resultsDF["model"] == m)][scorelist[0]], resultsDF[(resultsDF["nFeatures"] == plist[i]) & (resultsDF["Engineering"] == elist[i]) & (resultsDF["model"] == modelType[i])][scorelist[0]])[1] <=0.05):

                            p = plist[i]
                            a = alist[i]
                            e = elist[i]
                            m = modelType[i]
                elif plist[i] < p:
                    if alist[i] < a:
                        p = plist[i]
                        a = alist[i]
                        e = elist[i]
                        m = modelType[i]

            plist2.append(p)
            alist2.append(round(a,3))
            elist2.append(e)
            mlist2.append(m)
        
        namelist2.append(name)
        labelList2.append(label)
        featureElist2.append(elist2)
        parameterlist2.append(plist2)
        modelTypelist2.append(mlist2)
        averageslist2.append(alist2)
```

```python
newDF = pd.DataFrame()
newDF["data"] = namelist2
newDF["label"] = labelList2
newDF["average"] = averageslist2
newDF["model"] = modelTypelist2
newDF["selection"] = featureElist2
newDF["features"] = parameterlist2
```
